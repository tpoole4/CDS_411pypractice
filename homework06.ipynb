```python
def walk_hist_sim(nsteps):
    dist_list = []
    for trial in range(1000):
        walker_x = 0
        walker_y = 0
        walker_history = np.zeros((nsteps+1, 3))
        walker_history[:, 0] = np.arange(start=0, stop=nsteps+1, step=1)
        for step_number in np.arange(start=1, stop=nsteps + 1, step=1):
    # Generate the first random number rand for x coordinate
            rand = np.random.randint(low=0, high=2)
    # Conditional test
            if rand == 0:
                walker_x += 1
            else:
                walker_x -= 1

    # Generate the second random number rand for y coordinate
            rand = np.random.randint(low=0, high=2)
    # Conditional test
            if rand == 0:
                walker_y += 1
            else:
                walker_y -= 1
        
    # Store the step in the history
        walker_history[step_number, [1, 2]] = (walker_x, walker_y)
        walker_history = pd.DataFrame(data=walker_history, columns=["step", "x", "y"])
        walker_history["step"] = pd.to_numeric(walker_history["step"], downcast="signed")
        dist=np.sqrt(walker_x**2 + walker_y**2)
        dist_list.append(dist)
        av_dist=sum(dist_list)/len(dist_list)
    
    return(av_dist)
```

```python
walk_hist_sim(100)
```

```python
time_hist=[]

for walk_length in range(1,51):
    ave=walk_hist_sim(walk_length)
    time_hist.append([walk_length,ave])
```
